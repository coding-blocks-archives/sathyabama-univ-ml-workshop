```
#Import the OpenCV Library
import cv2

# Create a Face Detector Model
face_detector = cv2.CascadeClassifier('my_model.xml')
```

```
# Upload Photos
from google.colab import files
files.upload()
```

```
# Detect Faces in the Image Uploaded
# Load the Photo from Disk as 3D Matrix of Pixels

img = cv2.imread('my_image.jpg')


faces_detected = face_detector.detectMultiScale(img,1.3,5)

# List of Faces (Each Face is List of (x,y,w,h))
print(faces_detected)
```

```
# Import Google Colab Display Image Method 
from google.colab.patches import cv2_imshow


# Draw the Rectangles/Boundaries first
#Iterate over all rectangles detected
for f in faces_detected:
    x,y,w,h = f # f is of type (x,y,w,h)
    
    green_color = (0,255,255)
    cv2.rectangle(img,(x,y),(x+w,y+h),green_color,10)
    
# Use the method 'cv2_imshow' to show image
cv2_imshow(img)
```
