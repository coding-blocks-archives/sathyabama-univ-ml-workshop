```
# Problem 2

N = int(input())

# Iterate over Rows 1...N
for i in range(1,N+1):
    #First Print Spaces"
    print(" "*(N-i),end="")
    print("*"*(2*i-1))
    
    
```

```
# While Loop
i = 1
while i%7!=0:
    print(i)
    i = i + 1
```

```
# Problem 
# Print Factorial of a Number
N = int(input())

ans = 1
for i in range(1,N+1):
    ans = ans*i
    print(ans) #INside the Loop

print("Final Factorial "+str(ans)) #Outside the Loop
```

```
## Input a List of Numbers and Print their sum
# 1 5 9 8 10 12

numbers = [int(n) for n in input().split()]

ans = 0
for i in numbers:
    ans = ans + i
    
print(ans)
```

```
# Dictionary(Hashmap) in Python
price_table = {
    "Idli" : [20,40],
    "Dosa" : {"Onion":30,"Masala":40},
    "Sambhar":20,
}
```

```
#price_table['Idli'][1]
print(price_table['Dosa']['Onion'])
```

```
if price_table.get('Masala-Dosa') is None:
    print("Masala Dosa Not Present")
else:
    print(price_table['Masala-Dosa'])
    
```

```
# Insert New Items in the Price Table
price_table['Masala-Dosa'] = 50
print(price_table['Masala-Dosa'])
```

```
print(list(price_table.keys()))
print(list(price_table.values()))
```
